# Random Seed in PyTorch

A **seed** is the starting value used by the random number generator to start the sequence from the same location again and again without changing the values .


````python
import torch
````

````python
RANDOM_SEED = 23

torch.manual_seed(RANDOM_SEED)

torch.cuda.manual_seed(RANDOM_SEED)
````

````python
torch.rand(2)
````

# How to Accelerate Using GPU?

A GPU can perform many calculations at the same time, making deep learning much faster than a CPU.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

Move the model and data to the GPU:

```python
model = model.to(device)
X = X.to(device)
y = y.to(device)
```

**Note:** Model and data must be on the same device.

### Random Seed

```python
torch.manual_seed(42)
torch.cuda.manual_seed_all(42)
```

A seed is the starting point for generating random numbers.

* Same seed → Same results every run
* Different seed → Different results

# torch.nn

`nn` stands for **Neural Network**.

It provides ready-made building blocks for creating neural networks.

```python
import torch.nn as nn
```

Example:

```python
model = nn.Linear(3, 1)
```

Meaning:

* Takes 3 inputs
* Produces 1 output

PyTorch automatically creates and learns the weights and bias.

### Common nn Modules

* `nn.Linear()` → Fully connected layer
* `nn.ReLU()` → Activation function
* `nn.Sigmoid()` → Activation function
* `nn.Dropout()` → Reduces overfitting

**In simple words:** `torch.nn` is a toolbox containing ready-made parts for building neural networks.


````python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
````

````python
model = torch.nn.Linear(in_features=3, out_features=1)
````
