# Tensors Manipulation
here we will learn what operations we can perform on tensors 

```python
import torch
```

```python
#addition to tensors
tensors = torch.tensor([1,2,3,4,5])
tensors+10
```

```python
#multiplication to tensors
tensors*10
```

```python
#if two tensors are equal in size we can do element wise multiplication
tensors*tensors
```

```python
'''
we can also do matrix multiplication using @ and torch.matmul(A,B)
if A is of size (m,n) and B is of size (n,p) then the result will be of size (m,p)
and also the n and p should be equal for the multiplication to be possible
the "inner dimensions are consumed" in the multiplication
'''
```

```python
tensors.matmul(tensors)
# It will return dot product of the two tensors and the result will be a scalar value 
# because the two tensors are of size (5,) and (5,) and the result will be of size (1,)


''' 
Rule of thumb:

* 1D × 1D → scalar (dot product)
* 2D × 2D → matrix multiplication
* 2D × 1D → matrix-vector multiplication
* Higher dimensions → batched matrix multiplication
'''
```
